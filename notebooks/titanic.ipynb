```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('../data/titanic.csv')

# Remove rows with missing age
xf = df.dropna(subset=['age'])

# Create a violin plot passenger class and age

# Create a histogram of the passenger class and survival 
sns.violinplot(x='pclass', y='survived', data=xf)
plt.show()

 




 




```
